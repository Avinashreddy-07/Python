## __str__() is a special method in python that defines the human-readable string representation of an object.

```python
## Methods

class Person:
    def __init__(a, name, surname, year_of_birth):
        a.name=name
        a.surname = surname
        a.year_of_birth = year_of_birth
    def age(a, current_year):
        return current_year - a.year_of_birth
    def summ(a, current_year):
        return current_year + a.year_of_birth
    def __str__(a):
        return ("%s %s was born in %d."%(a.name,a.surname,a.year_of_birth))

alice = Person("Alice","Brown",1990)
print(alice)
print(alice.age(2026))
print(alice.summ(2026))
```

## __repr()__ is for developers - unambiguous, precise, ideally something that could be used to recreate the object.

```python
class Person:
    def __init__(self, name, surname, year_of_birth):
        self.name = name
        self.surname = surname
        self.year_of_birth = year_of_birth

    def __str__(self):
        return f"{self.name} {self.surname} was born in {self.year_of_birth}."

    def __repr__(self):
        return f"Person(name:'{self.name}', surname='{self.surname}', year_of_birth ={self.year_of_birth})"

alice = Person("Alice", "Brown", 1990)
print("Using str():", str(alice))
print("Using repr():", repr(alice))
```

## Abstraction: Hiding unnecessary details and exposing only what the user of your code actually needs.

```python
class Person:
    def __init__(self, name, surname, year_of_birth):
        self._name = name
        self._surname = surname
        self._year_of_birth = year_of_birth
        
    def age(self, current_year):
        return current_year - self._year_of_birth

    def __str__(self):
        return  f"{self._name} {self._surname} was born in {self._year_of_birth}"

alice = Person("Alice", "Brown", 1990)
print(alice)
print(alice._surname)
print(alice.age(2026))
```

# python's name-mangling rule
__ are mangled to prevent accidental access from outside the class.
print(alicee.__surname) -- it fails
 Because the real attribute name is _Person__surname
### To provide encapsulation — a core OOP principle
It ensures:
- You don’t accidentally override private attributes in subclasses
- You don’t accidentally access them from outside the class
It’s not true security, but it’s a safety mechanism.


```python
class Person:
    def __init__(self, name, surname, year_of_birth):
        self.__name = name
        self.__surname = surname
        self.__year_of_birth = year_of_birth
        
    def age(self, current_year):
        return current_year - self.__year_of_birth

    def __str__(self):
        return f"{self.__name} {self.__surname} was born in {self.__year_of_birth}"

alicee = Person("Alice","Brown",1990)
print(alicee)
print(alicee.age(2026))
print(alicee.__surname)
```

```python
class Person:
    def __init__(self, name, surname, year_of_birth):
        self.__name = name
        self.__surname = surname
        self.__year_of_birth = year_of_birth
    def get_name(self):
        return self.__name
    def get_surname(self):
        return self.__surname
    def get_year_of_birth(self):
        return self.__year_of_birth
    def age(self, current_year):
        return current_year - self.__year_of_birth
    def __str__(self):
        return "%s %s and was born in %d."%(self.__name, self.__surname, self.__year_of_birth)

alice = Person("Alice","Brown",1990)
print(alice.get_name())
print(alice.get_surname())
print(alice.get_year_of_birth())
print(alice.age(2026))
print(alice)
```

## Inheritance
The way one class can reuse, extend, or modify the behaviour of another.
Inheritance allows a class(child/subclass) to acquire the properties and methods of another class(Parent/superclass).

```python
# Basic Inheritance
## Dog class Inherits everything from Animal.
class Animal:
    def speak(self):
        return "Some sound"
class Dog(Animal):
    pass

d = Dog()
print(d.speak())
```

```python
# Adding or Overriding Methods
class Animal:
    def __init__(self, name):
        self.name = name   
    def speak(self):
        return "Some generic animal sound"
    def info(self):
        return f"Iam an animal named {self.name}"

class Dog(Animal):
    def speak(self):
        return "Bark"
    def fetch(self):
        return f"{self.name} is fetching the ball"

d = Dog("Bruno")
print(d)
print(d.speak())  # Overridden in Dog
print(d.info())   # Inherited from Animal
print(d.fetch())  # Defined only in Dog
```

```python
## 
class Person:
    def __init__(self, first_name, last_name, birth_year):
        self._first_name = first_name
        self._last_name = last_name
        self._birth_year = birth_year
        
class Student(Person):
    def __init__(self, student_id, *args):
        super().__init__(*args) #Person.__init__()
        self._student_id = student_id

c = Student(1,'charlie','Brown',1990)
print(c._student_id)
print(c._first_name)
```
