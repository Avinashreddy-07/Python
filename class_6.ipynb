### Errors and Exception Handling
##### Errors:  -  Those are the problems in your code that cannot be handled at runtime.
* Syntax Error: - wrong python syntax
* Indentation Error: - Incorrect Indentation
##### Exceptions( Runtime issues):  - Occur while the program is running and can be handled using try/except.
* ZeroDivisionError - dividing by zero.
* ValueError - wrong type of value.
* KeyError - missing dictionary key.
   
- Python allows you to catch these exceptions and prevent your program from crashing.

```python
try:
    f = open('testfile.txt','w')
    f.write('Test write this')
except IOError:
    print("Error: Could not find or read data")
else:
    print("Content written successfully")
    f.close()
```

```python
f = open('testfile.txt','r')
f.write('Test write this')
```

```python
try:
    f = open('testfile.txt','r')
    f.write('Test write this')
except IOError:
    # This will only check for an IOError exception and then execute this print statement
    print("Error: could not write in file or read data")
else:
    print("content written successfully")
    f.close()
```

```python
try:
    f = open("testfile.txt","w")
    f.write("Test write statement")
finally:
    print("Always execute finally code blocks")
```

```python
def askint():
    try:
        val = int(input("please enter an integer:"))
    except:
        print("Looks like you did not enter an integer")
    finally:
        print("Finally, I executed!")
```

```python
askint()
```

```python
def askint():
    try:
        val = int(input("Please enter an integer: "))
    except:
        print("Looks like you did not enter an Integer")
        try:
            val = int(input("Try again-please enter an integer:"))
        except:
            print("handle it")
    finally:
        print("Finally, I executed")
    print(val)
```

```python
askint()
```

```python
#seek
# r+ = open the file for reading and writing. Does not truncate(original stays)
with open("sample.txt","r+") as file:
    content = file.read()
    file.seek(35)
    file.write("chandler bing")
# file.seek(35) moves the file pointer to byte position 35.
# seek() + write = overwrite, not insert
```

```python
with open("sample.txt","r") as file:
    for line in file:
        print(line.strip())
```

```python
# If you want to insert
with open("sample.txt", "r+") as file:
    content = file.read()
    new_content = content[:35] + "chandler bing" + content[35:]
    file.seek(0)
    file.write(new_content)
    file.truncate()
```

### Working with Excel Files (.xlsx)

```python
import pandas as pd

#sample data
data = {
    "Name": ["Alice", "Bob", "Charlie", "Annie", "Raichel", "Armando"],
    "Marks": [85, 90, 78, 87, 11, 44]
}
df = pd.DataFrame(data)
df 
```

```python
# Writing to an Excel file
df.to_excel("students.xlsx",index=False)
## Reading from an Excel file
df_read = pd.read_excel("students.xlsx")
print("Data from Excel:")
print(df_read)
```

```python
### Database connectivity and operations using python.
import sqlite3
# connecting with the database.
db = sqlite3.connect("my_database3.db")

# Drop table if it already exist using execute() method.
db.execute("Drop table if exists grades1")

# Create table as per requirement
db.execute("create table grades1(id int, name text, score int)")

#Inserting values inside the created table
db.execute("Insert into grades1(id, name, score) values(101, 'John', 99)")
db.execute("Insert into grades1(id, name, score) values(102, 'Gary', 90)")
db.execute("Insert into grades1(id, name, score) values(103, 'James', 80)")
db.execute("Insert into grades1(id, name, score) values(104, 'Cathy', 85)")
db.execute("Insert into grades1(id, name, score) values(105, 'Anny', 95)")

# commit() some envi auto-commit, but its best practice to do it manually
db.commit()

# close() Ends connection safely
db.close()
```

```python
results = db.execute(" select * from grades1 order by id")
for row in results:
    print(row)
print("-" * 60)

results = db.execute("select * from grades1 where name = 'John' ")
for row in results: print(row)
print("-" * 60)
```
