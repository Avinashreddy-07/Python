### Iterators
An iterator is an object that lets you loop over data one item at a time.
* Must implement __iter__() -> returns the iterator object itself
* Must implement __next__() -> returns the next value
* Raises StopIteration  when no more items are left.
  ### Why iterators matter:
* Thay dont load everything into memory
* They allow lazy evaluation
* They power loops, comprehensions, file reading, generators, etc..

## Generators
Generator is a special type of Iterator created using:
* Generator functions (using yield)
* Generator expressions ( (x for x in range))
* Generators produce values lazily, one at a time, only when needed.

```python
# Generator function for the cube of numbers( power of 3)
def gencubes(n):
    for num in range(n):
        yield num**3
# Here yield is inside a function, that becomes a generator function.
```

```python
# Calling it does not run the function- it returns a generator object.
gencubes(10)
```

```python
for x in gencubes(9):
    print(x)
# Here in for loops it calls the next() until the generator is done. 
```

```python
for i in "ksdnsd":
    print(i)
```

```python
def genfibon(n):
    a = 1
    b = 1
    for i in range(n):
        yield a
        a,b = b, a+b
```

```python
genfibon(5) 
```

```python
for num in genfibon(12):
    print(num)
```

```python
def fibon(n):
    a = 1
    b = 1
    output = [] 

    for i in range(n):
        output.append(a)
        a,b = b,a+b
        
    return output
```

```python
fibon(10)
```

## next() and iter() built-in functions
* iter() takes an iterable( likes a list, string, tuple, range, or generator function) and returns an Iterator object.
* next() takes an iterator and returns the next item.
  

```python
# for loop uses iter() and next() internally.
for x in gencubes(5):
    print(x)

# Python Secretly does this:

gen = iter(gencubes(5)) #step-1
while True:
    try:
        x = next(gen) #step-2

        print(x)
    except StopIteration:
        break         #step-3
        
```

```python
def simple_gen():
    for x in range(3,6):
        yield x
```

## map()
map() applies a function to each item in a iterable( like a list, tuple, range) and returns a map object, which is lazy iterator.

```python
# map(function, iterable)
result = map(lambda x:x * 2, [1,2,3])
# Python stores the function lambda X:x*2
# python stores the list [1,2,3]
```

```python
next(result)  
```

### reduce()
reduce() repeatedly applies a function to the elements of a sequence, reducing the entire list into a single value.
* It works like this:
 - Take the first two elements → apply the function
 - Take the result → combine it with the next element
 - Continue until only one final value remain


```python
from functools import reduce
lst = ['Ram', 'Ramesh', 'venkatesh', 'suresh']
reduce(lambda a, b:a+b, lst)
```

```python
# Find the maximum of a sequence ( This already exists as max())
max_find = lambda a, b:a if (a > b) else b
```

```python
lst=[47,24,53,13,86]
```

```python
#find max
from functools import reduce
reduce(max_find, lst)
```

### filter
* If map() transforms items, reduce() collapses them, then filter() selects items based on a condition.
* filter(function, iterable) keeps only the elements for which the function returns True.
* It returns a lazy iterator, so you usually wrap it in list().

```python
# lets make a function
def even_check(num):
    if num%2 == 0:
        return True
```

```python
num = 9
even_check(num)
```

```python
lst = [1,2,3,4,5,6,7]

list(filter(even_check, lst))
```

```python
filter(even_check, lst)
```

* filter() is more commonly used in lambda functions, this because we usually use filter for a quick job where we dont want to write an entire function.

```python
list(map(lambda x: x%2 == 0, lst))
```

```python
words = ["Ram", "Ramesh", "Suresh", "Venu"]
long_words = list(filter(lambda word: len(word) >= 4, words))
```

```python
print(long_words)
```
