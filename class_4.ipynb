## Tuples
In python, tuples are similar to lists but they are immutable.
i.e they cannot be changed. 

you would use the tuples to present data that shouldn't be changed, such as days of weeks or dates on a calendar.

  1- constructing Tuples
  2- Basic Tuple Methods

#### Tuples 

* Ordered Collection
* Immutable (cannot change after creation)
* Allow duplicate values
* Can contain any data type
* Indexable ( you can do t[0])
* Hashable - if all elements are hashable -> can be used as dictionary keys
* Great for fixed, structured data.

* Use a tuple when the data should not change.
* Use it for fixed-size, structured records like coordinates 
* Choose tuples when you need faster and more memory-efficient alternatives to lists.
* Use tuples as dictionary keys or set elements boz they are hashable.

### Shallow copy vs Deep copy
- Shallow copy → copies the outer box, shares the inner items.( Creates a new container, but does not copy nested objects inside it.
  Inner Objects are shared between the original and the copy.)
- Deep copy → copies the outer box AND duplicates all inner items.(Created a new        container, and recursively copies all the nested objects.
  Changes in the copy do Not affect the Original.)

```python
# shallow copy
import copy
original = [1,2,3,4]
shallow = copy.copy(original)

shallow[0] =12

print('Original',original)
print('Shallow', shallow)
```

```python
# Deep copy
import copy
original = [[1,2],[3,4]]

deep = copy.deepcopy(original)

deep[0][0] = 12

print('Original', original)
print('Deep', deep)
```

### Introduction to Functions
Functions will be one of our main building blocks when we construct larger and larger amount of code to solve problems.

* A FUNCTION groups a set of statements together to run the statements more than once.
* Functions allow us to reuse the code instead of writing the code again and again.

```python
def print_num(n):
    for i in range(1,n+1):
        print(i*n)

print_num(5)
```

```python
def is_prime(num):
    for n in range(2,num):
        if num% n == 0:
            print('Not prime')
            break
    else :
        print('Prime')
```
