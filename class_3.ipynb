### List Comprehension

Python has a advanced feature called list comprehension which allows for Quick construction of lists.

```python
# Build a list comprehension by deconstucting a for loop within a []
first_col = [row[0] for row in matrix]
first_col
```

```python
first_col =[]
for row in matrix:
    first_col.append(row[0])
print(first_col)
```
