```python
# To Count the Vowels in the input string
def count_vowels(Input):
    vowels = 'aeiou'
    Input = Input.lower()
    count = 0
    for char in Input:
        if char in vowels:
            count = count+1
    return count

Input = input("Enter the String: ")
print(count_vowels(Input))
```

```python
# number is prime or not
def validation(number):
    if number > 1:
        for n in range(2, number):
            if number % n == 0:
                print("Not a prime number")
                break
        else:
            print("Prime Number")
    else:
        print("Not a prime number")
        
user_input = int(input("Enter the number:"))
validation(user_input)
```

```python
# print prime numbers upto N
def validation(number):
    if number == 2:
        print(number)
    if number > 2:
        for n in range(2, number+1):
            for i in range(2, n):
                if n % i == 0:
                    break
            else:
                print(n)

user_input = int(input("Enter the number:"))
validation(user_input)
```

```python
# print Fibonacci series.
def fib(number):
    a = 0
    b = 1
    for _ in range(number):
        print(a)
        a, b = b, b+a
        
number = int(input("Enter the number:"))
fib(number)
```

```python
# check if the number is Armstrong number or not

number = int(input("Enter the number:"))
digit = str(number)
power = len(digit)
total = 0
for i in digit:
    total += int(i) ** power
if total == number:
        print(number,"becomes",total, "Its Armstrong number")
else:
    print(number,"becomes",total, "Its not a Armstrong number")
```
